```python
import pandas as pd
df=pd.read_csv('/content/Book1.csv')
df
```

```python
target='pass or fail'
print(df[target])
```

```python
x=df.drop(target,axis=1)
print(x)
```

```python
w0=1
w1=3
w2=2
w3=4
w4=1
w5=2
z=w0*1+w1*x["study hours per week"]+w2*x["class attendance %"]+w3*x["previous exam score"]+w4*x["participation in group study"]+w5*x["subission of assignments"]
print(z)
yp=[]
for i in range(10):
  if z[i]>0:
    yp.append(1)
  else:
    yp.append(0)
```

```python
print(yp)
```
